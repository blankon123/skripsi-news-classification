```python
import requests
from bs4 import BeautifulSoup

import sys
import time
```

```python
linkcoba = 'http://detik.feedsportal.com/c/33613/f/656089/s/4e8048a9/sc/3/l/0Lfinance0Bdetik0N0Cread0C20A160C0A30C250C1120A170C31730A410C40Cakhir0Emaret0Einka0Eekspor0E150Ekereta0Emade0Ein0Emadiun0Eke0Ebangladesh/story01.htm'
```

```python
source_code = requests.get(linkcoba)
plain_text = source_code.text
soup = BeautifulSoup(plain_text)
soup
```

```python
def get_berita_detik(url):
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text)
    return soup.find(class_="text_detail")
```
